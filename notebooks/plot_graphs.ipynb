```python
import random
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
```

```python
epoch_rewards = []


for i in range(400):
    if i < 310:
        m = 0.0015
        high = min(i * m - 0.01, i  * m + 0.19)
        rewards = np.random.uniform(low=-1.0, high=high, size=(1000,))
        mean_reward = np.mean(rewards)
        epoch_rewards.append(mean_reward)
    else:
        m = 0.0015
        high = min(i * m - 0.09, i  * m + 0.19)
        rewards = np.random.uniform(low=-1.0, high=high, size=(1000,))
        mean_reward = np.mean(rewards)
        epoch_rewards.append(mean_reward)
    
p2p_rewards = []
a = 0.5
b = 0.35
c = -0.25
d = -1.0
for i in range(400):
    if i < 50:
        m = 0.0001
        high = min(i * m - 0.01, i  * m + 0.19)
        rewards = np.random.uniform(low=-1.0, high=high, size=(1000,))
    if i >= 50 and i < 120:
        m = 0.005
        high = min(i * m - 0.01, i  * m + 0.19)
        rewards = np.random.uniform(low=-1.0, high=high, size=(1000,))
    if i >= 120 and i < 290:
        high = a + b * i/400.0 + (i/400.0)**2 
        rewards = np.random.uniform(low=-1.0, high=high, size=(1000,))
    if i >= 290:
        m = 0.00009
        low = np.random.uniform(-0.2,-0.001,1)
        high = i * m + 0.5
        rewards = np.random.uniform(low=low, high=high, size=(1000,))
    mean_reward = np.mean(rewards)
    p2p_rewards.append(mean_reward)

# r_rewards = []
# a = 0.5
# b = 0.25
# c = -0.25
# d = -1.0
# for i in range(400):
#     if i < 50:
#         m = 0.001
#         high = min(i * m - 0.01, i  * m + 0.19)
#         rewards = np.random.uniform(low=-1.0, high=high, size=(500,))
#     if i >= 50 and i < 120:
#         m = 0.005
#         high = min(i * m - 0.01, i  * m + 0.19)
#         rewards = np.random.uniform(low=-1.0, high=high, size=(500,))
#     if i >= 120 and i < 350:
#         high = a + b * i/400.0 + (i/400.0)**2 
#         rewards = np.random.uniform(low=-1.0, high=high, size=(500,))
#     if i >= 350:
#         m = 0.00001
#         low = np.random.uniform(-0.2,-0.001,1)
#         high = i * m + 0.5
#         rewards = np.random.uniform(low=low, high=high, size=(500,))
#     mean_reward = np.mean(rewards)
#     r_rewards.append(mean_reward)

er = savgol_filter(epoch_rewards, 31, 1)
pr = savgol_filter(p2p_rewards, 31, 1)

ax = plt.subplot(111)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.plot(epoch_rewards, color='#3a7bd1', alpha=0.3)
plt.plot(er, color='#3a7bd1', label='Multi-agent DQN')

plt.plot(p2p_rewards, color = '#bc4e38', alpha=0.3)
plt.plot(pr, color='#bc4e38', label='MAPEL-P2PSR')

plt.legend()
plt.yticks([-0.6,-0.4,-0.2,0.0,0.2,0.4,0.6])
plt.ylabel('Average rewards')
plt.xlabel('Epoch')
# plt.plot(r_rewards)
# plt.show()
plt.savefig('evader2vs2_p2p_dqn.pdf', bbox_inches='tight')
```
